```python
%pylab inline
```

```python
import pandas as pd
```

```python
data =pd.read_table('output_all_scores.tab', index_col=0)
data.columns
```

# Observe the behaviour of the data 

```python
data.T.plot(legend=False, alpha=0.2)
xlabel('Metagenome length category (MSL)', weight='bold')
ylabel('SS', weight='bold')
plt.savefig('Score_data_behaviour1.png', dpi=300)
```

```python
m = data.mean()
s = data.std()

errorbar(m.index, m, yerr=s, fmt='k--o')

xlabel('Metagenome length category (MSL)', weight='bold')
ylabel(' SS', weight='bold')
plt.savefig('Score_data_behaviour2.png', dpi=300)
```

```python
ndata = data.T / data['30']
ndata
```

```python
ndata.plot(legend=False, alpha=0.2)
xlabel('Metagenome length category (MSL)', weight='bold')
ylabel('Normalized SS', weight='bold')
plt.savefig('Score_data_behaviour3.png', dpi=300)
```

# Plotting differential  ($x_i - x_{i-1}$)

```python
means = ndata.mean(1)
stds = ndata.std(1)
means
```

```python
errorbar(means.index, means, yerr = stds , fmt ='k--o')
xlabel('Metagenome length category', weight='bold')
ylabel('Normalized SS', weight='bold')
plt.savefig('Score_data_behaviour4.png', dpi=300)
```

```python
b = ndata.T.boxplot()
xlabel('Metagenome length category (MSL)', weight='bold')
ylabel('Normalized SS', weight='bold')
plt.savefig('Score_data_behaviour5.png', dpi=300)
```

```python
dif = ndata.diff()
dif.T.to_csv('diferential_data', sep='\t')
```

```python
dif.plot(alpha=0.1, legend=False)
xlabel('Metagenome length category (MSL)', weight='bold')
ylabel('Differential in SS', weight='bold')
plt.savefig('Score_data_behaviour6.png', dpi=300)
```

```python
dif_means = dif.mean(1)
dif_stds = dif.std(1)
errorbar(dif_means.index, dif_means, yerr=dif_stds, fmt='k--o')
xlabel('Mean size leght (MSL)', weight='bold')
ylabel('Differential in SS ($x_i - x_{i-1}$)', weight='bold')
savefig('diferencia_norm.png', dpi=300)
plt.savefig('Score_data_behaviour7.png', dpi=300)
```

```python
dif.T.boxplot()

xlabel('Metagenome length category', weight='bold')
ylabel('Diferential in SS', weight='bold')
plt.savefig('Score_data_behaviour8.png', dpi=300)
```
